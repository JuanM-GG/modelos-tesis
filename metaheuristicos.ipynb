```
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
```

```
def cicrcle(t):

    x = 2 + 2*np.cos(t) 
    y = 2 + 2*np.sin(t) 

    return x, y
```

```
t = np.linspace(0,10,1000)
x, y = cicrcle(t)
```

```
plt.figure(figsize= (7,7))
plt.plot((0,5),(0,5),"b--")
plt.plot((0,4),(6,0),"black")
plt.plot(x,y,"black")
plt.plot((12/5,3.5274),(12/5,0.7089),"r")
plt.ylabel("y", size = 15)
plt.xlabel("x", size = 15)
plt.legend([r"$y - x \neq 0$", "$y + 3x/2 - 6 = 0$",r"$(x-2)^2 + (y-2)^2 - 4 = 0$","espacio factible"], fontsize=12)
```

```
def convex(x):

    y = 1 + (x - 10)**2

    return y
```

```
x = np.linspace(5,15,100)
y = convex(x)
```

```
plt.figure(figsize=(8,8))
plt.plot(x,y)
x1 = 5.6
plt.scatter(x1,convex(x1), c = "red", s = 100)
x2 = 13
plt.scatter(x2,convex(x2), c = "red", s = 100)
plt.plot([x1, x2], [convex(x1), convex(x2)],"r--")
plt.text(6, 20, r'$f(x_1)$', fontsize = 15)
plt.text(13, 7.5, r'$f(x_2)$', fontsize = 15)
x_p = (x1 + x2)/2
convex_p = (convex(x1) + convex(x2))/2
plt.scatter(x_p,convex_p, c = "blue", s = 100) 
plt.text(x_p, convex_p+1, r'$\gamma f(x_1)+ (1- \gamma)f(x_2)$', fontsize = 15) 
plt.scatter(x_p,convex(x_p), c = "blue", s = 100) 
plt.text(x_p, convex(x_p)+1, r'$f(\gamma x_1 + (1- \gamma)x_2)$', fontsize = 15)
plt.ylabel("y", size = 15)
plt.xlabel("x", size = 15)
plt.text(8, 23, r'$f(x) = 1 + (x-10)^2$', fontsize = 20)
```
