```
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
```

```
def cicrcle(t):

    x = 2 + 2*np.cos(t) 
    y = 2 + 2*np.sin(t) 

    return x, y
```

```
t = np.linspace(0,10,1000)
xc, yc = cicrcle(t)
```

```
def convex(x):

    y = 1 + (x - 10)**2

    return y
```

```
x = np.linspace(5,15,100)
y = convex(x)
```

```
plt.figure(figsize=(8,8))
plt.ylim(-1,30)

plt.plot(x,y)
plt.plot(plt.xlim(), [0,0],"black")

x1 = 5.6
plt.scatter(x1,convex(x1), c = "red", s = 100)
plt.scatter(x1,0, c = "red", s = 100)

x2 = 13
plt.scatter(x2,convex(x2), c = "red", s = 100)
plt.scatter(x2,0, c = "red", s = 100)

plt.plot([x1, x2], [convex(x1), convex(x2)],"r--")
plt.text(6, 20, r'$f(x_1)$', fontsize = 15)
plt.text(13, 7.5, r'$f(x_2)$', fontsize = 15)

plt.plot([x1, x1], [0, convex(x1)],"g--")
plt.text(x1, 1, '$x_1$', fontsize = 15) 

plt.plot([x2, x2], [0, convex(x2)],"g--")
plt.text(x2, 1, '$x_2$', fontsize = 15) 

x_p = (x1 + x2)/2
convex_p = (convex(x1) + convex(x2))/2
plt.scatter(x_p,convex_p, c = "blue", s = 100) 
plt.text(x_p, convex_p+1, r'$\gamma f(x_1)+ (1- \gamma)f(x_2)$', fontsize = 15) 

plt.scatter(x_p,convex(x_p), c = "blue", s = 100) 
plt.text(x_p, convex(x_p)+1, r'$f(\gamma x_1 + (1- \gamma)x_2)$', fontsize = 15)

plt.ylabel("$f(x)$", size = 15)
plt.xlabel("$x$", size = 15)
plt.text(8, 23, r'$f(x) = 1 + (x-10)^2$', fontsize = 20)
```

```
def multimodal(x):

    y = -(x**2+x)*np.cos(x)

    return y
```

```
x = np.linspace(-10,10,100)
y = multimodal(x)
```

```
res1 = minimize(multimodal, -8, method='BFGS',
               options={'disp': True})
```

```
res1.x
```

```
res2 = minimize(multimodal, -2.5, method='BFGS',
               options={'disp': True})
```

```
res2.x
```

```
res3 = minimize(multimodal, 2.5, method='BFGS',
               options={'disp': True})
```

```
res3.x
```

```
res4 = minimize(multimodal, 7.5, method='BFGS',
               options={'disp': True})
```

```
res4.x
```

```
plt.figure(figsize= (7,7))

# graficar funciÃ³n
plt.plot(x,y)
plt.ylabel("$f(x)$", size = 15)
plt.xlabel("$x$", size = 15)

# graficar mÃ­nimos
# x1
plt.scatter(res1.x, multimodal(res1.x),c="red",s = 100)
# x2
plt.scatter(res2.x, multimodal(res2.x),c="red",s = 100)
# x3
plt.scatter(res3.x, multimodal(res3.x),c="red",s = 100)
# x4
plt.scatter(res4.x, multimodal(res4.x),c="red",s = 100)

# escribir la funciÃ³n
plt.text(-6, 80, r'$f(x) = -(x^2+x) \cos(x)$', fontsize = 20)


# dividir en intervalos 
plt.ylim(-150,150)
plt.plot([-10,-10],plt.ylim(),"b:")
plt.plot([res1.x, res1.x],plt.ylim(),"b:")
plt.plot([res2.x, res2.x],plt.ylim(),"b:")
plt.plot([res3.x, res3.x],plt.ylim(),"b:")
plt.plot([res4.x, res4.x],plt.ylim(),"b:")
plt.plot([10,10],plt.ylim(),"b:")

# escribir mÃ­nimos
plt.text(res1.x, multimodal(res1.x)-7, ("({x}, {f})".format(x=round(res1.x[0],3), f=round(multimodal(res1.x)[0],3))))
plt.text(res2.x-3, multimodal(res2.x)-11, ("({x}, {f})".format(x=round(res2.x[0],3), f=round(multimodal(res2.x)[0],3))))
plt.text(res3.x, multimodal(res3.x)-11, ("({x}, {f})".format(x=round(res3.x[0],3), f=round(multimodal(res3.x)[0],3))))
plt.text(res4.x+0.5, multimodal(res4.x), ("({x}, {f})".format(x=round(res4.x[0],3), f=round(multimodal(res4.x)[0],3))))

# intervalos 
plt.text((-10 + res1.x-1)/2, -120, "(I)", fontsize = 20)
plt.text((res1.x + res2.x-1.5)/2, -120, "(II)", fontsize = 20)
plt.text((res2.x + res3.x-2)/2, -120, "(III)", fontsize = 20)
plt.text((res3.x + res4.x -2)/2, -120, "(IV)", fontsize = 20)
plt.text((res4.x + 10-2)/2, -120, "(V)", fontsize = 20)
```

x_i = x_{i-1} - \lambda f'(x_{i-1})
$$

```
def dfunc(x):

    f = 2*x
    return f
```

```
def grad(x_a, alpha, maxit, eps):

    for i in range(maxit):

        print(x_a)

        x = x_a - alpha*dfunc(x_a)

        diff = abs(alpha*dfunc(x_a))

        if diff < eps:

            break

        x_a = x

    return x
```
