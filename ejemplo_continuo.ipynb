```
install.packages("deSolve")
library(deSolve)
```

Declarar la función

```
# parámetros 
k1 = 0.8877
k2 = 0.2320
k3 = 1.8260 
Caf = 100
```

```
objective <- function(D){

    P = (k1/(D+k2))*((-(D+k1) + sqrt((D+k1)^2 + 4*D*Caf*k3))/2*k3)
   

    return(P)
}
```

```
D = seq(0,1,0.01)
Pb = objective(D)
```

```
par(mar = c(5,7,4,2) + 0.1)
plot(D, Pb, type = "l", lwd = 4, cex.axis = 2, cex.lab = 2, col = "darkblue", xlab = "D (1/min)",
      ylab = expression(paste(P[C[B]]," (mol/L)")))
```

```
objetive_aux <- function(D) {

    f = - objective(D)

    return(f)
}
```

```
sol = optimise(objetive_aux, interval = c(0,1))
D_opt = sol$minimum
obj_opt = -sol$objective
```

```
par(mar = c(5,7,4,2) + 0.1)
plot(D, Pb, type = "l", lwd = 4, cex.axis = 2, cex.lab = 2, col = "darkblue", xlab = "D (1/min)",
      ylab = expression(paste(P[B]," (mol/L)")))
abline(v = D_opt, col = "brown", lwd = 4)
```
